#While Loop

```python
password = "abc123"
user_entry = input("Enter your password:")
print(user_entry == password)


if user_entry == password:
    print("Password is correct. Welcome to your account.")

else:
    attemps = 3
    while user_entry != password and attemps >0:
      print("Password is incorrect. Please try again.")

      user_entry = input("Enter your password:")
      if user_entry == password:
        print("Password is correct.")
        break
      attemps -=1
```

```python
#Create a while loop that counts from 0 to 10 and prints odd numbers to the screen.

number = 0

while number < 10:
  if number % 2 != 0:
    print(number)
  number +=1

```

```python
#Let's play a “guess the number game” using while loop.

original_number = 50
guess_number = int(input("Guess the number:"))


while guess_number != original_number:
  if guess_number < original_number:
    print("Little higher")
  else:
    print("Little lower")

  guess_number = int(input("Try again: "))

print("You have guessed the correct number. Congratulations!")


```
